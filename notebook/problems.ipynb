## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np

np.random.seed(7)
two_dices = np.random.randint(1,7,1000)+np.random.randint(1,7,1000)
repeat_count = []
for i in two_dices:
    if i > 7:
        repeat_count.append(True)
    else:
        repeat_count.append(False)

print(repeat_count)
print(f"The number of times a number greater than 7 appears:",(repeat_count.count(True)))
print(f"The probability of getting a number greater than 7 is:",(repeat_count.count(True)/1000*100),"%")

```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)

ball_box_number = []
for value in ball_box.values():
    if value == "White":
        ball_box_number.append(1)
    elif value == "Red":
        ball_box_number.append(2)
    else:
        ball_box_number.append(3)

print(f"The probability that the white ball comes out is:", ball_box_number.count(1)/60*100, "%")
print(f"The probability that the red ball comes out is:", ball_box_number.count(2)/60*100, "%")
print(f"The probability that the green ball comes out is:", ball_box_number.count(3)/60*100, "%")
            
```
